```python
import pandas as pd
```

### Heart Attack

```python
heart = pd.read_csv("../data/raw//heart.csv")

heart.Sex = heart.Sex.replace({'M':0, 'F':1})
heart.ExerciseAngina = heart.ExerciseAngina.replace({'N':0, 'Y':1})
heart.ST_Slope = heart.ST_Slope.replace({'Up':1, 'Flat':0, 'Down':-1})
heart.RestingECG = heart.RestingECG.replace({'ST':1, 'Normal':0, 'LVH':-1})
heart.ChestPainType = heart.ChestPainType.replace({'NAP':0, 'ASY':1, 'TA':2, 'ATA':3})

heart.head()
```

```python
heart.to_csv("../data/processed/heart.csv", index=None)
```

### Titanic

```python
titanic = pd.read_csv("../data/raw/Titanic-Dataset.csv")

titanic = titanic.drop(columns=["Cabin", "Name", "Ticket"], axis=1)
titanic.Fare = titanic.Fare.fillna(titanic.Fare.median())
titanic.Age = titanic.Age.fillna(titanic.Age.median())
titanic.Embarked = titanic.Embarked.fillna("C")
titanic.Sex = titanic.Sex.replace({'male':0, 'female':1})
titanic = pd.get_dummies(titanic)

titanic.head()
```

```python
titanic.to_csv("../data/processed/titanic.csv", index=None)
```
